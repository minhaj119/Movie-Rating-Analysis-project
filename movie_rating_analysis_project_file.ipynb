```python
# Movie Rating Analysis Project

# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Movie Rating Data
data = {
    'Movie_ID': [1, 2, 3, 4, 5, 6, 7, 8],
    'Title': ['Inception', 'Titanic', 'Avengers', 'Joker', 'Interstellar', 'Frozen', 'Coco', 'Dune'],
    'Genre': ['Sci-Fi', 'Romance', 'Action', 'Thriller', 'Sci-Fi', 'Animation', 'Animation', 'Sci-Fi'],
    'Release_Year': [2010, 1997, 2012, 2019, 2014, 2013, 2017, 2021],
    'IMDb_Rating': [8.8, 7.8, 8.0, 8.5, 8.6, 7.4, 8.4, 8.2],
    'Box_Office_Million': [830, 2187, 1519, 1074, 677, 1280, 807, 402]
}

# Creating DataFrame
df = pd.DataFrame(data)
print("‚úÖ Dataset created successfully!\n")
print(df)

# Save as CSV file (optional)
df.to_csv("movie_ratings.csv", index=False)
print("\nüìÅ movie_ratings.csv file saved successfully.")

# Check summary stats
print("\nüìä Summary statistics:")
print(df.describe())

# Average IMDb rating by Genre
avg_rating = df.groupby('Genre')['IMDb_Rating'].mean().sort_values(ascending=False)
print("\nüé¨ Average rating by genre:")
print(avg_rating)

# Plot 1: IMDb Rating by Genre
plt.figure(figsize=(8,5))
sns.barplot(x=avg_rating.index, y=avg_rating.values, palette='viridis')
plt.title("Average IMDb Rating by Genre")
plt.ylabel("Average IMDb Rating")
plt.xlabel("Genre")
plt.show()

# Plot 2: IMDb Rating vs Box Office
plt.figure(figsize=(8,5))
sns.scatterplot(x='Box_Office_Million', y='IMDb_Rating', hue='Genre', data=df, s=100)
plt.title("IMDb Rating vs Box Office Collection")
plt.xlabel("Box Office (in Millions)")
plt.ylabel("IMDb Rating")
plt.show()

print("\n‚úÖ Analysis completed successfully!")
```
